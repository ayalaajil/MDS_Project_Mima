```python
import pandas as pd
import numpy as np
import random
from autocorrect import Speller
import itertools
import re
import json
from metadata import language_registers, discussion_tones
import ast
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df_multi_poisson = pd.read_csv("dataset_generated_multiple_symptoms_per_phrase_poisson_distrib.csv")

df_multi_poisson['Symptoms'] = df_multi_poisson['Symptoms'].apply(ast.literal_eval)

# Compute the length of each list
df_multi_poisson['Symptom_Count'] = df_multi_poisson['Symptoms'].apply(len)


# Calculate percentage distribution
symptom_count_pct = (
    df_multi_poisson['Symptom_Count'].value_counts(normalize=True)
    .sort_index() * 100  # convert proportions to percentage
)

# Plot
ax = symptom_count_pct.plot(kind='bar')
plt.xlabel('Number of symptoms per phrase')
plt.ylabel('Percentage of phrases (%)')
plt.title('Distribution of symptom counts per phrase (in %)')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels on top of bars
for i, v in enumerate(symptom_count_pct):
    plt.text(i, v + 1, f"{v:.1f}%", ha='center', va='bottom', fontsize=9)

plt.ylim(0, symptom_count_pct.max() + 10)
plt.show()
```

```python
# Flatten the list of lists into a single list of symptoms
all_symptoms = [symptom for sublist in df_multi_poisson['Symptoms'] for symptom in sublist]

# Count frequencies
symptom_counts = Counter(all_symptoms)

# Convert to DataFrame for easier sorting/plotting
df_symptom_counts = pd.DataFrame(symptom_counts.items(), columns=['Symptom', 'Count']).sort_values(by='Count', ascending=False)

# Plot top N most frequent symptoms
top_n = 20
df_symptom_counts.head(top_n).set_index('Symptom').plot(kind='bar', legend=False)
plt.xlabel('Symptom')
plt.ylabel('Frequency')
plt.title(f'Top {top_n} most frequent symptoms')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
